## Data wrangling

### Subtask:
Prepare the data for spatial analysis by transforming the latitude and longitude coordinates and calculating distances between stops.


**Reasoning**:
Rename the 'X' and 'Y' columns to 'Longitude' and 'Latitude', calculate the Euclidean distance between every pair of stops, and find the minimum distance to any other stop for each stop.



```python
import numpy as np

# Rename columns
df_stops = df_stops.rename(columns={'X': 'Longitude', 'Y': 'Latitude'})

# Calculate Euclidean distance between every pair of stops
from scipy.spatial.distance import cdist

coordinates = df_stops[['Longitude', 'Latitude']].values
distance_matrix = cdist(coordinates, coordinates)
distance_df = pd.DataFrame(distance_matrix, index=df_stops['OBJECTID'], columns=df_stops['OBJECTID'])

# Add Distance to Nearest Stop
df_stops['DistanceToNearestStop'] = np.nanmin(distance_df, axis=1)

display(df_stops.head())
```

## Data clustering

### Subtask:
Perform spatial clustering on the LYNX Gold Line stops using the K-Means algorithm.


**Reasoning**:
Perform spatial clustering using K-Means algorithm on the longitude and latitude coordinates of the stops.

## Data analysis

### Subtask:
Analyze the clusters identified in the previous step to understand the characteristics of each cluster and identify potential service gaps.


**Reasoning**:
Calculate the average distance between stops within each cluster and the average distance between stops in different clusters. This will help identify clusters with a low density of stops or a large average distance between stops, indicating potential service gaps.



```python
# Calculate the average distance within each cluster
average_distance_within_cluster = []
for cluster_num in df_stops['cluster_label'].unique():
  cluster_stops = df_stops[df_stops['cluster_label'] == cluster_num]
  cluster_coordinates = cluster_stops[['Longitude', 'Latitude']].values
  if len(cluster_coordinates) > 1:
    cluster_distances = cdist(cluster_coordinates, cluster_coordinates)
    average_distance_within_cluster.append((cluster_num, np.mean(cluster_distances[cluster_distances > 0])))
  else:
    average_distance_within_cluster.append((cluster_num, 0))

# Calculate the average distance between clusters
average_distance_between_clusters = []
for cluster_num1 in df_stops['cluster_label'].unique():
  for cluster_num2 in df_stops['cluster_label'].unique():
    if cluster_num1 < cluster_num2:
      cluster1_stops = df_stops[df_stops['cluster_label'] == cluster_num1]
      cluster2_stops = df_stops[df_stops['cluster_label'] == cluster_num2]
      cluster1_coordinates = cluster1_stops[['Longitude', 'Latitude']].values
      cluster2_coordinates = cluster2_stops[['Longitude', 'Latitude']].values
      if len(cluster1_coordinates) > 0 and len(cluster2_coordinates) > 0:
        distances_between_clusters = cdist(cluster1_coordinates, cluster2_coordinates)
        average_distance_between_clusters.append((cluster_num1, cluster_num2, np.mean(distances_between_clusters)))

# Print the results
print("Average Distance Within Cluster:")
for cluster_num, avg_distance in average_distance_within_cluster:
  print(f"Cluster {cluster_num}: {avg_distance}")

print("\nAverage Distance Between Clusters:")
for cluster_num1, cluster_num2, avg_distance in average_distance_between_clusters:
  print(f"Between Cluster {cluster_num1} and Cluster {cluster_num2}: {avg_distance}")
```

**Reasoning**:
Identify clusters with a low density of stops or a large average distance between stops, indicating potential service gaps. Also, analyze the distribution of proposed and operating stops within each cluster to determine if there are clusters with a high concentration of proposed stops, signifying areas of planned future expansion.



```python
# Identify clusters with potential service gaps (low density or large average distance)
for cluster_num, avg_distance in average_distance_within_cluster:
  cluster_stops = df_stops[df_stops['cluster_label'] == cluster_num]
  if len(cluster_stops) < 3 or avg_distance > 4000:  # Define thresholds for low density and large distance
    print(f"Cluster {cluster_num} has potential service gaps:")
    print(f"Number of stops: {len(cluster_stops)}")
    print(f"Average distance between stops: {avg_distance}")

# Analyze the distribution of proposed and operating stops within each cluster
for cluster_num in df_stops['cluster_label'].unique():
  cluster_stops = df_stops[df_stops['cluster_label'] == cluster_num]
  proposed_stops = len(cluster_stops[cluster_stops['Status'] == 'Proposed'])
  operating_stops = len(cluster_stops[cluster_stops['Status'] == 'Operating'])
  if proposed_stops > operating_stops:
    print(f"\nCluster {cluster_num} has a high concentration of proposed stops, suggesting planned future expansion:")
    print(f"Proposed stops: {proposed_stops}")
    print(f"Operating stops: {operating_stops}")
```

## Feature engineering

### Subtask:
Create features that can be used to predict potential service gaps based on the existing data.


**Reasoning**:
Calculate the density of stops within a radius, the distance to the nearest proposed stop, and the average distance to stops within the same cluster. Then, based on these features, create a binary feature for "Service Gap Potential".



```python
# Calculate the Density of Stops within a Radius (e.g., 1 km)
radius_km = 1  # Define the radius for stop density calculation
from geopy.distance import geodesic

df_stops['StopDensity'] = 0
for index1, row1 in df_stops.iterrows():
  stop_count = 0
  for index2, row2 in df_stops.iterrows():
    if index1 != index2:
      distance = geodesic((row1['Latitude'], row1['Longitude']), (row2['Latitude'], row2['Longitude'])).km
      if distance <= radius_km:
        stop_count += 1
  df_stops.loc[index1, 'StopDensity'] = stop_count


# Calculate the Distance to the Nearest Proposed Stop
df_stops['DistanceToNearestProposed'] = float('inf')
for index1, row1 in df_stops[df_stops['Status'] == 'Operating'].iterrows():
  min_distance = float('inf')
  for index2, row2 in df_stops[df_stops['Status'] == 'Proposed'].iterrows():
    distance = geodesic((row1['Latitude'], row1['Longitude']), (row2['Latitude'], row2['Longitude'])).km
    if distance < min_distance:
      min_distance = distance
  df_stops.loc[index1, 'DistanceToNearestProposed'] = min_distance


# Calculate the Average Distance to Stops within the Same Cluster
df_stops['AverageDistanceWithinCluster'] = 0
for cluster_label in df_stops['cluster_label'].unique():
  cluster_stops = df_stops[df_stops['cluster_label'] == cluster_label]
  if len(cluster_stops) > 1:
    cluster_coordinates = cluster_stops[['Longitude', 'Latitude']].values
    cluster_distances = cdist(cluster_coordinates, cluster_coordinates)
    average_distance = np.mean(cluster_distances[cluster_distances > 0])
    df_stops.loc[df_stops['cluster_label'] == cluster_label, 'AverageDistanceWithinCluster'] = average_distance


# Create a Binary Feature for "Service Gap Potential"
df_stops['ServiceGapPotential'] = 0
for index, row in df_stops.iterrows():
  if row['StopDensity'] < 2 and row['DistanceToNearestProposed'] > 2 and row['AverageDistanceWithinCluster'] > 3:
    df_stops.loc[index, 'ServiceGapPotential'] = 1

display(df_stops.head())
```

## Model training

### Subtask:
Train a classification model to predict the likelihood of a service gap based on the engineered features.


**Reasoning**:
Prepare the data for model training by splitting it into features (X) and the target variable (y), and then splitting it into training and testing sets.



```python
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LogisticRegression

# Separate features and target variable
X = df_stops[['StopDensity', 'DistanceToNearestProposed', 'AverageDistanceWithinCluster']]
y = df_stops['ServiceGapPotential']

# Handle infinite values (replace with NaN and then impute)
X = X.replace([float('inf'), float('-inf')], np.nan)
X = X.fillna(X.mean())

# Split the data into training and testing sets
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.25, random_state=42)

# Train a Logistic Regression model
model = LogisticRegression()
model.fit(X_train, y_train)

# Make predictions on the test set
y_pred = model.predict(X_test)
```

## Model evaluation

### Subtask:
Evaluate the performance of the trained Logistic Regression model.


**Reasoning**:
Evaluate the trained Logistic Regression model using metrics like accuracy, precision, recall, F1-score, and confusion matrix.



```python
from sklearn.metrics import accuracy_score, precision_score, recall_score, f1_score, confusion_matrix

# Calculate evaluation metrics
accuracy = accuracy_score(y_test, y_pred)
precision = precision_score(y_test, y_pred)
recall = recall_score(y_test, y_pred)
f1 = f1_score(y_test, y_pred)

print(f"Accuracy: {accuracy}")
print(f"Precision: {precision}")
print(f"Recall: {recall}")
print(f"F1-score: {f1}")

# Generate and print the confusion matrix
cm = confusion_matrix(y_test, y_pred)
print("\nConfusion Matrix:")
print(cm)
```
